```python
#Transactions is going to have some correlation with sales data

transaction_sales=pd.merge(train.groupby(['store_nbr','date']).sales.sum().reset_index(),transactions,how="left")
transaction_sales.head(5)
```

```python
#find out spearman correlation
print("Spearman Correlation between Total Sales and Transactions: {:,.4f}"
      .format(transaction_sales.corr("spearman").sales.loc["transactions"]))
```
